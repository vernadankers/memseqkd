```python
sns.set_theme(style="whitegrid")
results = pd.read_csv("results/quality.tsv", sep='\t')
results = results[results['model'].isin(['S', 'T', 'B'])]

P = ["#90509B", "#587AB9", "#35ADB1"]
fig, axes = plt.subplots(1, 2, figsize=(7, 2))
[ax1, ax4] = axes
langpair_to_x = {'En-De': 0, 'De-En': 1, 'Pl-En': 2, "En-Pl": 3, "Fr-De": 4}

def create_subplot(results, metric, ax, label):
    x, y, h = [], [], []
    for i, _ in enumerate(results["langpair"]):
        xcoord = langpair_to_x[results["langpair"].tolist()[i]]

        y.append(results[metric].tolist()[i])
        h.append(results["model"].tolist()[i])
        if 'T' in h[-1]:
            xcoord -= 0.1
        elif 'B' in h[-1]:
            xcoord += 0.1
        x.append(xcoord)
        
    sns.scatterplot(x=x, y=y, hue=h, style=h,
                  palette=P, markers=['o', 'v', '^'], ax=ax,  s=100)
    ax.set_xticks([0, 1, 2, 3, 4], langs)
    ax.set_ylabel(label)
    ax.set_xlabel("")
    ax.xaxis.grid(False)

create_subplot(results, "BLEU", ax1, "BLEU"+ r"$\uparrow$")
ax1.set_ylim(25, 45)
create_subplot(results, "Comet-QE-22", ax4, "Comet-QE-22"+ r"$\uparrow$")
ax4.set_ylim(75, 86)

ax4.legend([],[], frameon=False)
ax4.set_xticks([0, 1, 2, 3, 4], langs)
plt.tight_layout(pad=1.5)
handles, labels = ax1.get_legend_handles_labels()
ax1.legend(handles, [r"$\theta_T$", r"$\theta_S$", r"$\theta_B$"], fontsize=9)
plt.savefig("figures/model_quality_bars_short.pdf", bbox_inches="tight")
```

### Section 2.2, percentual increase of memorization/hallucination metrics

```python
results = pd.read_csv("results/memorization.tsv", sep='\t')
results = results[results['model'].isin(['T', 'S', 'B'])]
for metric, label in [('replication_tc', 'replication'),
                      ('exmem_tc', 'exmem'),
                      ("oschal", "oschal"),
                      ("nathal", "nathal")]:
    langpairs = results[results['model'] == 'S']["langpair"]
    em_teacher = results[results['model'] == 'T'][metric]
    em_student = results[results['model'] == 'S'][metric]
    em_baseline = results[results['model'] == 'B'][metric]
    increase = (em_student.values - em_baseline.values) / em_baseline.values * 100
    print(f"Percentual increase S over B: {np.mean(increase):.2f}+/-{np.std(increase):.2f}")

    plt.figure(figsize=(3 if metric != "oschal" else 2.6, 1.2))
    ax = sns.barplot(y=increase, x=langpairs.values, color=P[1],  saturation=1, width=0.4)
    plt.ylabel("% increase", fontsize=10)
    sns.despine()
    plt.xticks([])
    if metric in ["oschal", "exmem_tc"]:
        ax.set_yticks([0, 30, 60])
    plt.savefig(f"figures/{label}_increase_barplot.pdf", bbox_inches="tight")
    plt.show()
```

### Appendix D, Adaptive-SeqKD

```python
palette = {l: c for c, l in 
           zip(sns.color_palette("tab20", 5), ["En-De", "De-En", "Pl-En", "En-Pl", "Fr-De"])}
palette["corpus"] = "grey"

table = ""
for fn, metric, name in [
        ("results/quality", "BLEU", "BLEU"),
        ("results/quality", "Comet-22", "Comet-22"),
        ("results/quality", "Comet-QE-22", "Comet-QE-22"),
        ("results/memorization", "exmem_tc", "ExMem " + r"$\mathcal{T}_C$"),
        ("results/memorization", "exmem_tt", "ExMem " + r"$\mathcal{T}_T$"),
        ("results/memorization", "oschal", "OscHal"),
        ("results/memorization", "oschal_cc", "OscHal (CC)"),
        ("results/memorization", "nathal", "NatHal")]:
    results = pd.read_csv(f"{fn}.tsv", sep='\t')
    plt.figure(figsize=(2.75 if metric != "exmem_tt" else 1.25, 2.5))
    x, y, h = [], [], []
    res = defaultdict(list)
    for langpair in ["En-De", "De-En", "Pl-En", "En-Pl", "Fr-De"]:
        student1 = results[(results['langpair'] == langpair) & (results['model'] == "S")]
        teacher1 = results[(results['langpair'] == langpair) & (results['model'] == "T")]

        student2 = results[(results['langpair'] == langpair) & (results['model'] == "S_ft-hq")]
        teacher2 = results[(results['langpair'] == langpair) & (results['model'] == "T_ft-hq")]

        student3 = results[(results['langpair'] == langpair) & (results['model'] == "S_ft-ra")]
        teacher3 = results[(results['langpair'] == langpair) & (results['model'] == "T_ft-ra")]

        if metric != "exmem_tt":
            a = (teacher2[metric].values[0] - teacher1[metric].values[0])/teacher1[metric].values[0]*100
            b = (teacher3[metric].values[0] - teacher1[metric].values[0])/teacher1[metric].values[0]*100
            res[metric, r"$T_{hq}$"].append(a)
            res[metric, r"$T_{ra}$"].append(b)
            x.extend([r"$T_{hq}$", r"$T_{ra}$"])
            y.extend([a, b])
            h.extend([langpair, langpair])

        c = (student2[metric].values[0] - student1[metric].values[0])/student1[metric].values[0]*100
        d = (student3[metric].values[0] - student1[metric].values[0])/student1[metric].values[0]*100

        res[metric, r"$S_{hq}$"].append(c)
        res[metric, r"$S_{ra}$"].append(d)
        
        x.extend([r"$S_{hq}$", r"$S_{ra}$"])
        y.extend([c, d])
        h.extend([langpair, langpair])

    sns.barplot(x=x, y=y, hue=h, palette=palette, width=0.5)
    plt.ylabel(f"% {name.replace('Comet', 'Com')} diff. to $T$ / $S$")
    if metric != "oschal_cc":
        plt.legend([],[], frameon=False)
    else:
        plt.legend(bbox_to_anchor=(1, 1), fontsize=9)
    plt.savefig(f"figures/{metric.lower()}_difference_adaptiveseqkd.pdf", bbox_inches="tight")
    
    for x in res:
        table += f"& {np.mean(res[x]):.1f}$\\pm${np.std(res[x]):.1f}"
    table += '\\\\\n'
```
