```python
import pickle
import os
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from collections import defaultdict
sns.set_context("talk")
sns.set_theme(style="whitegrid")

model_names = [("teacher",  r"$\theta_T$"),
               ("student_b=1_seed=1111", r"$\theta_S$"),
               ("baseline_seed=1111",  r"$\theta_B$")]
palette_models = ["#90509B", "#587AB9", "#35ADB1"]
```

### Appendix E.2, random subgroup

```python
results = defaultdict(list)
x, x_models, cometqe, replication, chrf, msttr, comet, lang, lang2 = [], [], [], [], [], [], [], [], []
for langpair in ["en-de", "de-en", "pl-en", "en-pl", "fr-de"]:
    for subgroup_name in ["random.pickle"]:
        subgroup = pickle.load(open(f"{langpair}/" + subgroup_name, 'rb'))
        for model1, model2 in [("corpus", r"$\mathcal{T}_C$")] + model_names:
            x.append(model2)
            cometqe.append(subgroup[model1]["comet22-qe"] * 100)
            msttr.append(subgroup[model1]["msttr"])
            comet.append(subgroup[model1]["comet22"] * 100)
            lang.append(langpair)
            if model1 != "corpus":
                x_models.append(model2)
                replication.append(subgroup[model1]["replicate_corpus"])
                chrf.append(subgroup[model1]["chrf"])
                lang2.append(langpair)

palette = {l: c for c, l in 
           zip(sns.color_palette("tab20", 5),
               ["en-de", "de-en", "pl-en", "en-pl", "fr-de"])}
palette["corpus"] = "grey"

fig, [ax0, ax1, ax2, ax3] = plt.subplots(1, 4, figsize=(10, 2.5))
ax = sns.scatterplot(x=x_models, y=replication, hue=lang2, palette=palette,
                     ax=ax0, s=50)
ax = sns.lineplot(x=x_models, y=replication, color="black", marker='s',
                  linestyle='', legend=False, err_style='bars', ax=ax0, alpha=0.5, errorbar='sd')
ax.legend([],[], frameon=False)
ax.set_ylabel("% replicated from "+ r"$T_{C}$")

ax = sns.scatterplot(x=x_models, y=chrf, hue=lang2, palette=palette,
                     legend=False, ax=ax1, s=50)
ax = sns.lineplot(x=x_models, y=chrf, color="black", marker='s', linestyle='',
                  legend=False, err_style='bars', ax=ax1, alpha=0.5, errorbar='sd')
ax.set_ylim(54, 70)
ax.set_xlim(-0.5, 2.5)
ax.legend([],[], frameon=False)
ax.set_ylabel("chrF")

ax = sns.scatterplot(x=x, y=cometqe, hue=lang, palette=palette, ax=ax2, s=50)
ax = sns.lineplot(x=x, y=cometqe, color="black", marker='s', linestyle='',
                  legend=False, err_style='bars', ax=ax2, alpha=0.5, errorbar='sd')
ax.set_ylim(65, 85)
ax.legend([],[], frameon=False)
ax.set_ylabel("Comet-QE-22")

ax = sns.scatterplot(x=x, y=msttr, hue=lang, palette=palette, legend=False,
                     ax=ax3, s=50)
ax = sns.lineplot(x=x, y=msttr, color="black", marker='s', linestyle='',
                  legend=False, err_style='bars', ax=ax3, alpha=0.5, errorbar='sd')
ax.set_ylim(0.65, 0.8)
ax.set_xlim(-0.5, 3.5)
plt.ylabel("MSTTR")

handles, labels = ax2.get_legend_handles_labels()
ax3.legend(handles, labels, frameon=True, fontsize=9, bbox_to_anchor=(1, 1.05))

plt.tight_layout()
plt.savefig("figures/random_subgroup.pdf", bbox_inches="tight")
plt.show()
```

### Appendix E.2, CM, Comet, confidence subgroups

```python
def label(ax, group):
    if group == "cm":
        ax.set_xticks(
            [0, 1, 2, 3, 4, 5, 6],
            ["L-L", "0.4+", "0.3-0.4", "0.2-0.3", "R", "0-0.2", "H-H"], rotation=90)   
    elif group == "confidence":
        ax.set_xticks(
            [0, 1, 2,],
            ["C" + r"$\downarrow$", "R", "C" + r"$\uparrow$"], rotation=90)   
    else:
        ax.set_xticks(
            [0, 1, 2, 3, 4, 5],
            ["0-0.2", "0.2-0.4", "0.4-0.6", "R", "0.6-0.8", "0.8-1"], rotation=90)

for group in ["cm", "confidence", "comet"]:
    models, x = [], []
    metrics = defaultdict(list)

    # Predefine xpositions for nicely formatted and readable graphs
    if group == "cm":
        buckets = [("cm_bottomleft", 0), ("cm_0.4-1", 1), ("cm_0.3-0.4", 2),
                   ("cm_0.2-0.3", 3), ("random", 4), ("cm_0-0.2", 5), ("cm_topright", 6)]
    elif group == "confidence":
        buckets = [("low_confidence", 0), ("random", 1), ("high_confidence", 2) ]
    else:
        buckets = [("comet22_0-0.2", 0), ("comet22_0.2-0.4", 1), ("comet22_0.4-0.6", 2),
                   ("random", 3), ("comet22_0.6-0.8", 4), ("comet22_0.8-1", 5)]

    for subgroup_name, x_value in buckets:
        for langpair in ["en-de", "de-en", "en-pl", "pl-en", "fr-de"]:
            subgroup = pickle.load(open(f"{langpair}/{subgroup_name}.pickle", 'rb'))
            for model1, model2 in model_names:
                models.append(model2)
                for metric in ["comet22", "comet22-qe", "replicate_corpus", "chrf", "msttr"]:
                    metrics[metric].append(subgroup[model1][metric])

                # Slightly offset teacher & baseline such that they don't overlap in graphs
                x.append(x_value - (0.2 if "teacher" in model1 else 0) + \
                         (0.2 if "baseline" in model1 else 0))

    fig, [ax1, ax2, ax3, ax4, ax5] = plt.subplots(
        1, 5, figsize=(11 if group!= "cm" else 12, 2.5 if group != "confidence" else 2.1))
    
    def plot_metric(y, ax, ylabel):
        ax = sns.lineplot(x=x, y=y, color="black", marker='o', legend=False,
                          hue=models, linestyle='', err_style='bars', errorbar='sd', ax=ax,
                          palette=palette_models)
        ax.set_ylabel(ylabel)
        label(ax, group)
    plot_metric(metrics["replicate_corpus"], ax1, "% replicated " + r"$\mathcal{T}_C$")
    plot_metric(metrics["chrf"], ax2, "chrF " + r"$\mathcal{T}_C$")
    plot_metric(metrics["comet22"], ax3, "Comet-22")
    plot_metric(metrics["comet22-qe"], ax4, "Comet-QE-22")
    plot_metric(metrics["msttr"], ax5, "MSTTR")
    plt.legend([], [], frameon=False)
    plt.tight_layout()
    plt.savefig(f"figures/{group}_subgroups.pdf", bbox_inches="tight")
    plt.show()
```

### Appendix E.2, summarize percentual change in Comet-QE-22 scores

```python

def plot_rel_diff(buckets, name, size):
    results = defaultdict(list)
    metric = "comet22-qe"
    for langpair in ["fr-de", "en-de", "pl-en", "en-pl", "de-en"]:
        for subgroup_name in buckets:
            subgroup = pickle.load(open(f"{langpair}/{subgroup_name}.pickle", 'rb'))     
            student_name = "student_b=1_seed=1111"
            baseline_name = "baseline_seed=1111"

            results["exceeds_teacher"].append(
                (subgroup[student_name][metric] - subgroup["teacher"][metric]) / \
                    subgroup["teacher"][metric] * 100)
            results["baseline_exceeds_teacher"].append(
                (subgroup[baseline_name][metric] - subgroup["teacher"][metric]) / \
                    subgroup["teacher"][metric] * 100)

            # Pretty-format the subgroups' names for readability            
            results["subgroup"].append(subgroup_name.replace(".pickle", "").replace(
                "comet22_", "Qua=").replace("cm_", "CM=").replace("_", "").replace(
                "topright", "H-H").replace("bottomleft", "L-L").replace(
                "lowconfidence", r"conf$\downarrow$").replace("highconfidence",
                r"conf$\uparrow$").replace("0.", ".")
            )

    # Two bar graphs, one for student one for baseline
    results = pd.DataFrame(results)
    fig, ax1 = plt.subplots(1, figsize=size)
    ax = sns.barplot(
        results.sort_values("exceeds_teacher", ascending=False), y="baseline_exceeds_teacher", x="subgroup",
        label=r"$\theta_B$", ax=ax1, alpha=1 if metric != "chrf" else 0.7, linewidth=2, edgecolor="#35ADB1",
        color="#35ADB1", fill=True, zorder=-1 if metric == "chrf" else 10)
    ax = sns.barplot(
        results.sort_values("exceeds_teacher", ascending=False), y="exceeds_teacher", x="subgroup", label=r"$\theta_S$",
        ax=ax1, linewidth=2, edgecolor="#587AB9", alpha=0.5, fill=True, color="#587AB9")

    # Remove axes, re-add horizontal lines
    sns.despine(top=True, right=True, left=True, bottom=True)
    ax.yaxis.grid(False)
    ax.xaxis.grid(False)
    for x_ in [0, 2.5, 5, 7.5, 10, 12.5, 15]:
         sns.lineplot(y=[x_, x_], x=[0, len(set(results["subgroup"]))-1], ax=ax1, color='grey', zorder=-1)
    h = ax.get_xticks()
    l = ax.get_xticklabels()
    ax.set_xticks(h, l, rotation=35)
    ax.tick_params(axis='x', pad=0)
    plt.ylabel("Comet-QE-22\ndifference (%)")
    plt.xlabel("")
    plt.legend(bbox_to_anchor=(0.85, 0.7))
    plt.savefig(f"figures/subgroups_{metric}_{name}.pdf", bbox_inches="tight")

buckets = [
    "random", "cm_0-0.2", "cm_0.2-0.3", "cm_0.3-0.4", "cm_0.4-1",
    "comet22_0-0.2", "comet22_0.2-0.4", "comet22_0.4-0.6", "comet22_0.6-0.8", "comet22_0.8-1",
    "cm_bottomleft", "cm_topright", "low_confidence", "high_confidence"]
plot_rel_diff(buckets, "large", (8, 2))

buckets = [
    "random", "cm_0-0.2", "cm_0.2-0.3", "cm_0.3-0.4", "cm_0.4-1",
    "comet22_0-0.2", "comet22_0.2-0.4", "comet22_0.4-0.6", "comet22_0.6-0.8", "comet22_0.8-1",
    "low_confidence", "high_confidence"]
plot_rel_diff(buckets, "small", (8, 2))
```
